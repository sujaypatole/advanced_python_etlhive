### Multiply by 3 inside list vs array

```python
a = [5,6,7,8,9,10,11]
type(a)
```

```python
# We cant multiply all list element by 3
a*3 # This repeats a list 3 times
```

```python
b = []
for i in a:
   b.append(i*3)
```

```python
b
```

```python
# This is using list comprehension
# You will have to use the for loop for multiplication 
# no matter whichever method you use
c = [i*3 for i in a] # List comprehension
```

```python
c
```

```python
a1 = np.array(a)
a1
```

```python
a1*3
```

```python
# Broadcasting numpy
3*a1 + 5 
# Operation gets applied on to all of your array elements in single application. You can perform any mathematical 
# Operation on numpy array. This is called as numpy broadcasting
# Also numpy i svery faster than list
```
